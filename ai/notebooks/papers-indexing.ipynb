```python
def PIL2OpenCV (pil_image):
    open_cv_image = numpy.array(pil_image) 
    # Convert RGB to BGR 
#     return open_cv_image[:, :, ::-1].copy() 
    return cv2.cvtColor(numpy.array(pil_image), cv2.COLOR_RGB2BGR)
```
